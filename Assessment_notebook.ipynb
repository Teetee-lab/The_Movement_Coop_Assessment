```python
#concatenate the firstname, middlename and last name
#if there is no middle name fill the nan with a space
All_data['name'] = All_data['FIRST_NAME'].str.cat(All_data.MIDDLE_NAME.fillna(""),sep=" ").str.cat(All_data.LAST_NAME,sep=" ").str.replace(' ',' ')
All_data.head(10)
```

```python
#drop the lastname, firstname, and middlename since we already join them and put it in a new column
All_data = All_data.drop(['LAST_NAME', 'FIRST_NAME','MIDDLE_NAME','SUFFIX'], axis =1)
All_data.head(10)
```

```python
#convert the date of birth to datetime data type
#create a new column with just the year
#then drop the Date of birth column, now that we have new column that we need
#view the first ten rows 
All_data['DATE_OF_BIRTH']=pd.to_datetime(All_data['DATE_OF_BIRTH'])
All_data['birth_year'] = All_data['DATE_OF_BIRTH'].dt.year
All_data = All_data.drop('DATE_OF_BIRTH', axis =1)
All_data.head(10)
```

```python
#convert the residential zip to an int data type
#create a new column name zip, then drop the residential zip
All_data['zip'] = All_data['RESIDENTIAL_ZIP'].astype('Int64')
All_data = All_data.drop('RESIDENTIAL_ZIP',axis =1)
All_data.head(10)
```

```python
#rename the columns we will be needing to match input data
master_data = All_data.rename(columns = {'SOS_VOTERID':'voters_id','COUNTY_NUMBER': 'county_number',
                                  'COUNTY_ID':'county_id','RESIDENTIAL_ADDRESS1': 'address','RESIDENTIAL_CITY': 'city',	
                                  'RESIDENTIAL_STATE':'state'})
master_data.head(10)
```

```python
#change the string name, address and city by capitalizing the Initials
#view the first ten rows
master_data['name']= master_data['name'].str.title()
master_data['address']= master_data['address'].str.title()
master_data['city']= master_data['city'].str.title()

master_data.head(10)
```
